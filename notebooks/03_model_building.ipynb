# Step 3: Build LSTM Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
model = Sequential()

model.add(LSTM(50, activation='relu', input_shape=(10,1)))
model.add(Dense(1))
```

```python
model.compile(
    optimizer="adam",
    loss="mse"
)
```
